```python
# NAME: SHER ALI
# ROLL NO: P21-8024
# BCS-6

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

df=pd.read_csv("fruit_data_with_colors _1_.csv")

text_columns = ["fruit_name", "fruit_subtype"]
# dropping the unwanted columnss
df = df.drop(text_columns, axis=1)

# here find mean and filling the missing values in the date set
imputer = SimpleImputer(strategy="mean")
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

print (df_imputed)

X = df_imputed.drop("fruit_label", axis=1)
y = df_imputed["fruit_label"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


k_values = range(1, 11)
accuracy_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    accuracy = knn.score(X_test, y_test)
    accuracy_scores.append(accuracy)
    # print(f"Accuracy for K={k}: {accuracy}")

# Step f: Plot the accuracy score for each value of K
plt.plot(k_values, accuracy_scores, marker='o')
plt.title('Accuracy Score vs. K Value')
plt.xlabel('K Value')
plt.ylabel('Accuracy Score')
plt.xticks(k_values)
plt.grid(True)
# plt.show()
```
