```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

#Reading data from dataframe
df=pd.read_csv('/home/sherali/Downloads/dataset.csv')

#This will fill the missing values in the columns according to the mean for each coloumn
df.fillna(df.mean(),inplace=True)

#Printing the Data frame to check for the missing values
print(df)

G=nx.Graph()

#Adding nodes from the SOurce and Target Data
G.add_nodes_from(df['source'],age=df['age'],connections=df['connections'])
G.add_nodes_from(df['target'],age=df['age'],connections=df['connections'])

#Adding Edges 
G.add_edges_from(zip(df['source'],df['target']))

nx.draw(G,with_labels=True)
```

```python
import matplotlib.pyplot as plt

#Generating X_Values from 0 to 20
x_values=np.linspace(0,20,5)

#First FUnction Y=2X+1
y1=(2*x_values)+1

#Second FUnction COS(X)
y2=np.cos(x_values)


#plottting , with labels and using markers
plt.plot(x_values,y1, label=' Y = 2x+1 ', marker='o')
plt.plot(x_values,y2, label=' cosx ', marker='x')



plt.xlabel('X')
plt.ylabel('Y')

plt.legend()
plt.show()

```
